#Situation und Aufgabe

Im Fach EMFF dürfen wir zwei freiwählbare Semesterarbeiten schreiben. In der Arbeit von Mark Wildi, analysieren wir den Zusammenhang zwischen dem SMI und verschiedenen Währungskursen und versuchen den SMI anhand der Währungskurse vorherzusagen. 
Das Ziel dieser Arbeit ist herauszufinden, ob es Zusammenhänge zwischen dem S&P500 und verschiedenen Währungen gibt. Die USA exportiert jährlich Waren und Dienstleistungen im Wert von rund \$1'600 Milliarden und importiert Waren rund \$2'500 Milliarden. Dieser Handel wird nicht immer in der Landeswährung also in US-Dollar durchgeführt, sondern auch in anderen Währungen.

Unsere Fragestellung lässt sich einfacher mit einem Beispiel erklären. Dieses Beispiel gilt nicht nur für den Importhandel, sondern auch für den Exporthandel:

Eine amerikanische Firma importiert Güter, die in Schwizer Franken gehandelt werden. Der Preis für CHF 1 beträgt heute USD 0.90. Wir nehmen an, dass das Gut X pro Stück CHF 100 kostet. Die Firma bezahlt für jedes Erworbene Stück USD 90.Einen Monat später befindet sich der Wechselkurs bei für USD/CHF bei USD 1.10. Die Schweizer Firma verkauft immer noch dasselbe Gut X welches auf dem Markt für CHF 100 erhältlich ist. Die amerikanische Firma muss für das gleiche Gut X pro Stück jetzt USD 110 bezahlen. Weil das Gut X nicht in US-Dollar sondern in CHF gehandelt wird, muss die amerikanische Firme USD 20 mehr bezahlen. 
Natürlich beeinflussen diese Preisunterschiede die Gewinne der amerikanischen Firma. Dieses Beispiel ist stets ein Extrembeispiel, um unsere Fragestellung deutlich zu veranschaulichen.

Um den S&P500 Kurs mit Hilfe der Währungskurse vorherzusagen und Zusammenhänge zu erkennen verwenden wir verschiedene Methoden. Wir verwenden Decision Trees, Random Forests, Support Vector Machine und Neuronale Netze.
